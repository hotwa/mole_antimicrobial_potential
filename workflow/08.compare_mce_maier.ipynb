# Comparing the MCE and Maier chemical libraries

```python
import os
# Data Manipulation
import numpy as np
import pandas as pd

# Data Viz
import seaborn as sns
import matplotlib.pyplot as plt

# Low dimensional embeddings
import umap
```

## Prepare directories

```python
RAW_DATA_DIR = "../raw_data"

MAIER_DIR = "../data/01.prepare_training_data/"
MCE_DIR = "../data/04.new_predictions/"

OUTPUT_DIR = "../data/08.compare_mce_maier"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

## Read Data

```python
# MolE representations

maier_mole = pd.read_csv(os.path.join(MAIER_DIR, "maier_mole_representation.tsv.gz"), sep='\t', index_col=0)
mce_mole = pd.read_csv(os.path.join(MCE_DIR, "MolE_representation_medchemexpress.tsv.gz"), sep='\t', index_col=0)

# Fingerprint representations
maier_fps = pd.read_csv(os.path.join(MAIER_DIR, "maier_ecfp4_representation.tsv.gz"), sep='\t', index_col=0)
mce_fps = pd.read_csv(os.path.join(MCE_DIR, "ecfp4_representation_medchemexpress.tsv.gz"), sep='\t', index_col=0)
```

```python
## Metadata
mce_metadata = pd.read_excel(os.path.join(MCE_DIR, "mole_mce_predictions.xlsx"), sheet_name="mole_prediction_overview")
maier_metadata = pd.read_excel(os.path.join(RAW_DATA_DIR, "maier_microbiome", "chem_library_info_SF1.xlsx"))

# Show the molecules used for training
maier_split = pd.read_csv(os.path.join(MAIER_DIR, "maier_scaffold_split.tsv.gz"), sep='\t')
maier_train = maier_split.loc[maier_split["split"]=="train", "prestwick_ID"].unique()
```

```python
# Gather MolE representations
mce_mole = mce_mole.loc[mce_metadata["Catalog Number"].unique()]
maier_mole = maier_mole.loc[maier_train]


# ECFP4 representations
mce_fps = mce_fps.loc[mce_metadata["Catalog Number"].unique()]
maier_fps = maier_fps.loc[maier_train]
```

## UMAP embeddings

```python
# MolE representation
combined_mole = pd.concat([mce_mole, maier_mole])

reducer = umap.UMAP(random_state=42, n_components=2, n_neighbors=25, min_dist=0.3, metric="cosine")

embedding = reducer.fit_transform(combined_mole.values)

mole_umap = pd.DataFrame(data = embedding,
                          columns = ['umap1', 'umap2'], 
                          index = combined_mole.index)
                          
# Add chemical library origin
mole_umap = mole_umap.reset_index(names="chem_id")

mole_umap["Chemical Library"] = mole_umap["chem_id"].apply(lambda x: "MCE" if x in mce_metadata["Catalog Number"].unique() else "Maier et.al.")
mole_umap.head()
```

```python
sns.scatterplot(mole_umap, x="umap1", y="umap2", s=10, hue="Chemical Library",
                palette={"MCE": "#C5C5C5", "Maier et.al.": "#DE1F84"})
sns.despine(bottom=True, left=True)
plt.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
#plt.legend(frameon=False)
```

```python
## Ecfp4 representation
combined_fps = pd.concat([mce_fps, maier_fps])

reducer = umap.UMAP(random_state=42, n_components=2, n_neighbors=25, min_dist=0.3, metric="jaccard")

embedding = reducer.fit_transform(combined_fps.values)

fps_umap = pd.DataFrame(data = embedding,
                          columns = ['umap1', 'umap2'], 
                          index = combined_fps.index)
                          
# Add mol weight information
fps_umap = fps_umap.reset_index(names="chem_id")
fps_umap["Chemical Library"] = fps_umap["chem_id"].apply(lambda x: "MCE" if x in mce_metadata["Catalog Number"].unique() else "Maier et.al.")
fps_umap.head()
```

```python
sns.scatterplot(fps_umap, x="umap1", y="umap2", s=10, hue="Chemical Library",
                palette={"MCE": "#C5C5C5", "Maier et.al.": "#DE1F84"})
sns.despine(bottom=True, left=True)
plt.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
```

## Compare experimentally validated molecules to the Maier training set. 

```python
from scipy.spatial.distance import pdist, squareform, cdist
```

Compare the chosen molecules to all compounds used for model training.

```python
# Gather the ECPF4 representation for the chosen chemicals
chosen_fps = combined_fps.loc[list(chosen_chems.keys())]

# Calculate the pairwise distances with respect to Maier et al. compounds
fps_distance = pd.DataFrame(cdist(chosen_fps, maier_fps, metric="jaccard"),
                            index = chosen_fps.index, columns=maier_fps.index)

fps_distance = fps_distance.rename(index=chosen_chems)
```

```python
fps_distance.min(axis=1)
```

Compare chosen compounds to antibiotics present in the Maier et al. dataset

```python
bacteria_targeted_maier = maier_metadata.loc[(maier_metadata["target species"] == "bacteria") &
                                             (maier_metadata["prestwick_ID"].isin(maier_fps.index)), "prestwick_ID"].to_list()
abx_maier_fps = maier_fps.loc[bacteria_targeted_maier]
```

```python
# Calculate the pairwise distances with respect to Maier et al. compounds that are known antibiotics
fps_distance_abx = pd.DataFrame(cdist(chosen_fps, abx_maier_fps, metric="jaccard"),
                            index = chosen_fps.index, columns=abx_maier_fps.index)

fps_distance_abx = fps_distance_abx.rename(index=chosen_chems)
fps_distance_abx.min(axis=1)
```

Compare to the non antibiotic compounds with broad-spectrum activity in Maier training 

```python
SCREEN_DATA_DIR = "../data/01.prepare_training_data/"
screen_df = pd.read_csv(os.path.join(SCREEN_DATA_DIR, "maier_screening_results.tsv.gz"), index_col="prestwick_ID", sep='\t')

screen_counts = screen_df.sum(axis=1).to_frame("inhib_count")
screen_counts.head()
```

```python
# Gather the IDs of the non-antibiotics
non_abx_maier = maier_metadata.loc[(maier_metadata["target species"] != "bacteria") &
                                             (maier_metadata["prestwick_ID"].isin(maier_fps.index)), "prestwick_ID"].to_list()

# Determine the non-antibiotics that have a broad spectrum activity
nonabx_bsa_maier = screen_counts.loc[non_abx_maier].loc[screen_counts["inhib_count"] > 10].index

# Gather the ECPF4 representation for the non-antibiotics with BSA
nonabx_maier_fps = maier_fps.loc[nonabx_bsa_maier]


# Calculate the pairwise distances with respect to Maier et al. compounds that are known antibiotics
fps_distance_nonabx = pd.DataFrame(cdist(chosen_fps, nonabx_maier_fps, metric="jaccard"),
                            index = chosen_fps.index, columns=nonabx_maier_fps.index)

fps_distance_nonabx = fps_distance_nonabx.rename(index=chosen_chems)
fps_distance_nonabx.min(axis=1)
```
